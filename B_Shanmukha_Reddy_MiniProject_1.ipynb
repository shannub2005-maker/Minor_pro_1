```python
import pandas as pd
import re

# Read WhatsApp chat file
with open("hostel_bois.txt", "r", encoding="utf-8") as f:
    chat = f.read()
pattern = r"(\d{1,2}/\d{1,2}/\d{2,4}),.*? - (.*?): (.*)"
messages = re.findall(pattern, chat)

df=pd.DataFrame(messages,columns=["Date","User","Message"])
print("Total messages:",len(df))
# Mesasages by user
print("\nMessages by user:")
print(df["User"].value_counts())
#Most active member
print("\n Most active member:")
print(df["User"].value_counts().idxmax())
#Least active member
print("\n least active member:")
print(df["User"].value_counts().idxmin())
```

```python
members = set()

for msg in messages:
    members.add(msg[1])

print("Total Members:", len(members))
print(members)
```

```python
from collections import Counter
import re
users = []
for line in chat.split('\n'):
  match = re.search(r", (\d{1,2}):\d{2} - (.*?):", line)
  if match:
    hour = int(match.group(1))
    user = match.group(2).strip()
    if hour >= 22 or hour <= 4:
      users.append(user)

if users:
  print("Night owl members:")
  most_common_night_owl = Counter(users).most_common(1)[0][0]
  print(most_common_night_owl)
else:
  print("No night activity found")
```

```python
import pandas as pd
import re
from collections import Counter


full_pattern = r"(\d{1,2}/\d{1,2}/\d{2,4}), (\d{1,2}:\d{2}) - (.*?): (.*)"
full_messages_data = re.findall(full_pattern, chat)

df_report = pd.DataFrame(full_messages_data, columns=["Date_str", "Time_str", "User", "Message"])


df_report["DateTime"] = pd.to_datetime(df_report["Date_str"] + " " + df_report["Time_str"], format='%d/%m/%y %H:%M')


total_messages = len(df_report)


members = set(df_report["User"].unique())


start_date = df_report["DateTime"].min()
end_date = df_report["DateTime"].max()


df_report['Date_only'] = df_report['DateTime'].dt.date
max_day = df_report['Date_only'].value_counts().idxmax()


df_report['Hour'] = df_report['DateTime'].dt.hour
busy_hour = df_report['Hour'].value_counts().idxmax()


person_messages = df_report['User'].value_counts().to_dict()

word_count = {}
average_length = {}
for person in members:
    user_messages_df = df_report[df_report['User'] == person]


    all_words_for_person = " ".join(user_messages_df["Message"].astype(str)).split()
    word_count[person] = len(all_words_for_person)


    if len(user_messages_df) > 0:
        total_word_count_in_messages = sum(len(msg.split()) for msg in user_messages_df["Message"].astype(str))
        average_length[person] = total_word_count_in_messages / len(user_messages_df)
    else:
        average_length[person] = 0

print("\n")
print("=" * 60)
print("                  GROUPDNA REPORT")
print("=" * 60)


print("\nGENERAL OVERVIEW")
print("-" * 60)

print(f"Total Messages : {total_messages}")
print(f"Total Members  : {len(members)}")
print(f"Start Date     : {start_date.date()}")
print(f"End Date       : {end_date.date()}")
print(f"Duration       : {(end_date - start_date).days} Days")
print(f"Busiest Day    : {max_day}")
print(f"Busiest Hour   : {busy_hour}")


print("\n" + "=" * 60)
print("MEMBER STATISTICS")
print("=" * 60)

for person in sorted(members):

    messages_count = person_messages.get(person, 0)
    words = word_count.get(person, 0)
    avg_len = average_length.get(person, 0)

    contribution = (messages_count / total_messages) * 100 if total_messages > 0 else 0

    print(f"\nName            : {person}")
    print(f"Messages        : {messages_count}")
    print(f"Words           : {words}")
    print(f"Average Length  : {avg_len:.2f}")
    print(f"Contribution    : {contribution:.2f}%")
    print("-" * 60)
```
